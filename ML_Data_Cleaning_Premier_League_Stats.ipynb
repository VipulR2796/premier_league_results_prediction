Importing required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from matplotlib import pyplot
```

```python
epl_data = pd.read_csv("https://raw.githubusercontent.com/VipulR2796/premier_league_results_prediction/main/df_full_premierleague.csv")
```

```python
epl_data.info(verbose=True)
```

```python
epl_data['home'] = epl_data['result_full'].str.split('-').str[0]
epl_data['Away'] = epl_data['result_full'].str.split('-').str[1]
epl_data.head()
```

```python
epl_data['Winner'] = epl_data[['home', 'Away']].apply(lambda row: "Home" if row['home'] > row['Away'] else "Away" if row['home'] < row['Away'] else "Draw" if row['home'])
```
